Used Chatgpt to help

```python
# Numeric columns (exclude target)
#select_dtypes(include=["int64", "float64"])
# selects all numeric columns

#.columns
# extracts just the column names

#.drop("SalePrice")
# removes the target variable so we do not alter the outcome variable
num_cols = df.select_dtypes(include=["int64", "float64"]).columns.drop("SalePrice")

#Loops through each numeric column
#Replaces missing values (NaN) with the median of that column

for c in num_cols:
    df[c] = df[c].fillna(df[c].median())


# Categorical columns
#Selects all categorical variables stored as text (object)
#Extracts their column names
cat_cols = df.select_dtypes(include="object").columns

#Replaces missing category values with an explicit label "Missing"
for c in cat_cols:
    df[c] = df[c].fillna("Missing")
```

```python
# This loop converts categorical columns from generic text (object) to pandas category data type.
for c in cat_cols:
    df[c] = df[c].astype("category")
```

```python
# This code reduces the influence of extreme outliers in key size-related numeric variables by capping their values at the 1st and 99th percentiles.
for c in ["LotArea", "GrLivArea", "TotalBsmtSF"]:
    lower = df[c].quantile(0.01)
    upper = df[c].quantile(0.99)
    df[c] = df[c].clip(lower=lower, upper=upper)
```
